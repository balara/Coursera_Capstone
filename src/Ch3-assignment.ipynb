<strong> Mappig postal codes to lattitude and longitude </strong>
<ul>
    <li> Read the data from CSV file</li>
    <li> Merge the data from CSV to the web data</li>
    <li> drop all NAN values i.e for which the lattitude and longitude not available in the csv file.-</li> 
</ul>


```python
ldf = pd.read_csv('Geospatial_Coordinates.csv')
merged_df = pd.merge(codes[0], ldf, 
                     left_on = 'Postal Code', 
                     right_on = 'Postal Code', 
                     how='outer')

merged_df = merged_df.dropna()
merged_df.sort_values(by=['Postal Code'])
merged_df.head(20)

```

<strong>Plot the Toronto downton</strong>


```python
latitude=43.654260
longitude=-79.360636


postal_map = folium.Map(location=[latitude, longitude], zoom_start=15) # generate map centred around Downtown Toronto


# add Ecco as a red circle mark
folium.CircleMarker(
    [latitude, longitude],
    radius=50,
    popup='Downtown Toronto',
    fill=True,
    color='red',
    fill_color='red',
    fill_opacity=0.5
   ).add_to(postal_map)


# display map
postal_map
```

```python
filtered_df = merged_df[merged_df["Borough"].str.contains("Toronto")]
print("Total Cout", merged_df.shape)
print("Toronto Neighbourgoods count", filtered_df.shape)
```

```python

# add popular spots to the map as blue circle markers
for lat, lng, label in zip(filtered_df['Latitude'], filtered_df['Longitude'], filtered_df['Postal Code']):
    folium.CircleMarker(
        [lat, lng],
        radius=10,
        popup=label,
        fill=True,
        color='blue',
        fill_color='blue',
        fill_opacity=0.5
        ).add_to(postal_map)
 
postal_map
```
